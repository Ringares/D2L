```python
import torch
import numpy as np
import sys; sys.path.insert(0, '../')
import d2l_utils
%matplotlib inline
```

```python
fair_probs = [1.0 / 6] * 6
counts = np.random.multinomial(10, fair_probs, size=500) 
cum_counts = counts.astype(np.float32).cumsum(axis=0) 
estimates = cum_counts / cum_counts.sum(axis=1, keepdims=True) 
d2l_utils.set_figsize((6, 4.5)) 
for i in range(6): 
    d2l_utils.plt.plot(estimates[:, i], label=("P(die=" + str(i + 1) + ")")) 
d2l_utils.plt.axhline(y=0.167, color='black', linestyle='dashed')

d2l_utils.plt.gca().set_xlabel('Groups of experiments') 
d2l_utils.plt.gca().set_ylabel('Estimated probability') 
d2l_utils.plt.legend();
```
